# `smolagents`

## What is smolagents?

`smolagents` is an open-source Python library designed to make it extremely easy to build and run agents using just a few lines of code.

Key features of `smolagents` include:

âœ¨ **Simplicity**: The logic for agents fits in ~thousand lines of code. We kept abstractions to their minimal shape above raw code!

ğŸ§‘â€ğŸ’» **First-class support for Code Agents**: [`CodeAgent`](https://huggingface.co/docs/smolagents/main/en/reference/agents#smolagents.CodeAgent) writes its actions in code (as opposed to "agents being used to write code") to invoke tools or perform computations, enabling natural composability (function nesting, loops, conditionals). To make it secure, we support [executing in sandboxed environment](https://huggingface.co/docs/smolagents/main/en/tutorials/secure_code_execution) via [Modal](https://modal.com/), [Blaxel](https://blaxel.ai), [E2B](https://e2b.dev/), or Docker.

ğŸ“¡ **Common Tool-Calling Agent Support**: In addition to CodeAgents, [`ToolCallingAgent`](https://huggingface.co/docs/smolagents/main/en/reference/agents#smolagents.ToolCallingAgent) supports usual JSON/text-based tool-calling for scenarios where that paradigm is preferred.

ğŸ¤— **Hub integrations**: Seamlessly share and load agents and tools to/from the Hub as Gradio Spaces.

ğŸŒ **Model-agnostic**: Easily integrate any large language model (LLM), whether it's hosted on the Hub via [Inference providers](https://huggingface.co/docs/inference-providers/index), accessed via APIs such as OpenAI, Anthropic, or many others via LiteLLM integration, or run locally using Transformers or Ollama. Powering an agent with your preferred LLM is straightforward and flexible.

ğŸ‘ï¸ **Modality-agnostic**: Beyond text, agents can handle vision, video, and audio inputs, broadening the range of possible applications. Check out [this tutorial](https://huggingface.co/docs/smolagents/main/en/examples/web_browser) for vision.

ğŸ› ï¸ **Tool-agnostic**: You can use tools from any [MCP server](https://huggingface.co/docs/smolagents/main/en/reference/tools#smolagents.ToolCollection.from_mcp), from [LangChain](https://huggingface.co/docs/smolagents/main/en/reference/tools#smolagents.Tool.from_langchain), you can even use a [Hub Space](https://huggingface.co/docs/smolagents/main/en/reference/tools#smolagents.Tool.from_space) as a tool.

ğŸ’» **CLI Tools**: Comes with command-line utilities (smolagent, webagent) for quickly running agents without writing boilerplate code.

### Create Your First Agent

````python
from smolagents import CodeAgent, InferenceClientModel

# Initialize a model (using Hugging Face Inference API)
model = InferenceClientModel()  # Uses a default model

# Create an agent with no tools
agent = CodeAgent(tools=[], model=model)

# Run the agent with a task
result = agent.run("Calculate the sum of numbers from 1 to 10")
print(result)
````

````python
from smolagents import CodeAgent, InferenceClientModel

# Initialize a model (using Hugging Face Inference API)
model = InferenceClientModel()  # Uses a default model

# Create an agent with no tools
agent = CodeAgent(tools=[], model=model)

# Run the agent with a task
result = agent.run("You have three boxes: Box A contains two gold coins (GG), Box B contains two silver coins (SS), Box C contains one gold and one silver coin (GS),You choose a box uniformly at random, then draw one coin at random from that box and observe that the coin is gold.Question: What is the probability that the other coin in the same box is also gold?")
print(result)
````

That's it! Your agent will use Python code to solve the task and return the result.

# **1. DuckDuckGoSearchTool**

````python
from smolagents import CodeAgent, InferenceClientModel, DuckDuckGoSearchTool

model = InferenceClientModel()
agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    model=model,
)

# Now the agent can search the web!
result = agent.run("What is the current weather in Paris?")
print(result)
````

````python
from smolagents import CodeAgent, InferenceClientModel, DuckDuckGoSearchTool

model = InferenceClientModel()
agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    model=model,
)

# Now the agent can search the web!
result = agent.run("Who won the ICC Women's World Cup 2025?")
print(result)
````

DuckDuckGoSearchTool often returns limited or incomplete search snippets, not full articles, causing the agent to miss crucial facts.
Because of this, it may give incorrect or outdated answers when the required information isn't clearly present in the snippet results.
**Truth**:**India won the ICC Women's World Cup 2025 .They beat South Africa by 52 runs.**

**Fix The Agent**

**1. Using better model**

````python
from smolagents import CodeAgent, InferenceClientModel, DuckDuckGoSearchTool

model = InferenceClientModel(
    model_id="meta-llama/Llama-3.1-70B-Instruct"
)

agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    model=model,
)

result = agent.run("Who won the ICC Women's World Cup 2025?")
print(result)
````

**2. Using recursive search prompt**

````python
from smolagents import CodeAgent, InferenceClientModel, DuckDuckGoSearchTool

model = InferenceClientModel(
    provider="openai",
    api_key=os.environ["OPENAI_API_KEY"],
    model_id="gpt-4o-mini"   # Must be a string
)

agent = CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    model=model,
)

result = agent.run("""
Search again until you find a reliable, true and grounded answer from an authorized news source.
(Operation: recursively refine search query)
Question: Who won the ICC Women's World Cup 2025?
""")

print(result)
````
